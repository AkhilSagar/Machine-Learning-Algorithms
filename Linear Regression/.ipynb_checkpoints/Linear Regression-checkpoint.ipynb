# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Admission_Prediction.csv')
data.head()
```

```python
data.shape
```

```python
#print the summary of dataframe
data.info()
```

```python
#finding the count of missing values from different columns
data.isna().sum()
```

```python
data.describe()
```

```python
# filling nan values
data['University Rating'].fillna(data['University Rating'].mode()[0], inplace=True)
data['TOEFL Score'].fillna(data['TOEFL Score'].mean(),inplace = True)
data['GRE Score'].fillna(data['GRE Score'].mean(), inplace = True)
```

```python
data.describe()
```

```python
data = data.drop(columns=['Serial No.'])
data.head()
```

```python
#lets viualize the data
import seaborn as sns
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1
for col in data:
    if plotnumber<=16:
        ax = plt.subplot(4,4,plotnumber)
        sns.distplot(data[col])
        plt.xlabel(col,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
y = data['Chance of Admit']
X =data.drop(columns = ['Chance of Admit'])
```

```python
plt.figure(figsize=(20,30), facecolor='white')
plotnumber = 1

for column in X:
    if plotnumber<=15 :
        ax = plt.subplot(5,3,plotnumber)
        plt.scatter(X[column],y)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Chance of Admit',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_scaled
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables,i) for i in range(variables.shape[1])]
vif['features'] = X.columns
```

```python
vif
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=100)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
y_pred = regression.predict(x_test)
y_pred
```

```python
regression.predict([[300,110,5,5,5,10,1]])
```

```python
from sklearn.metrics import r2_score
score = r2_score(regression.predict(x_test),y_test)
score
```

```python
# saving the model to the local file system
import pickle
filename = 'finalized_model.pickle'
pickle.dump(regression, open(filename, 'wb'))
```

```python
# prediction using the saved model
loaded_model = pickle.load(open(filename, 'rb'))
a=loaded_model.predict(scaler.transform([[300,110,5,5,5,10,1]]))
a
```

```python
regression.score(x_train,y_train)
```

```python
# Let's create a function to create adjusted R-Squared
def adj_r2(x,y):
    r2 = regression.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adj_r2(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```

```python
adj_r2(x_test,y_test)
```

## Lasso Regularization

```python
from sklearn.linear_model  import Ridge,Lasso,RidgeCV, LassoCV, ElasticNet, ElasticNetCV
# LassoCV will return best alpha and coefficients after performing 10 cross validations
lasscv = LassoCV(alphas = None,cv =10, max_iter = 10000, normalize = True)
lasscv.fit(x_train, y_train)
```

```python
# best alpha parameter
alpha = lasscv.alpha_
alpha
```

```python
#now that we have best parameter, let's use Lasso regression and see how well our data has fitted before
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

## Ridge regression model

```python
alphas = np.random.uniform(low=0, high=10, size=(50,))
ridgecv = RidgeCV(alphas = alphas,cv=10,normalize = True)
ridgecv.fit(x_train, y_train)
```

```python
ridgecv.alpha_
```

```python
# ridge_model = Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(x_train, y_train)
```

```python
ridge_model.score(x_test, y_test)
```

## Elastic net

```python
elasticCV = ElasticNetCV(alphas = None, cv =10)
elasticCV.fit(x_train, y_train)
```

```python
elasticCV.alpha_
```

```python
elasticCV.l1_ratio
```

```python
elasticnet_reg = ElasticNet(alpha = elasticCV.alpha_,l1_ratio=0.5)
elasticnet_reg.fit(x_train, y_train)
```

```python
elasticnet_reg.score(x_test, y_test)
```
